### 과제 2)
- '뽐뿌' 커뮤니티 내 아래 검색 메뉴에서 '제목만'으로 11번가 검색 '후' 상위 게시글 '번호' 3개를 출력

```python
import requests
from bs4 import BeautifulSoup

url = 'http://www.ppomppu.co.kr/zboard/zboard.php?id=ppomppu'
r = requests.get(url, {'keyword' : '11번가', 'search_type' : '제목만'})
html = r.text

soup = BeautifulSoup(html, 'lxml')
soup

odd_post_nums = soup.select('.list0 > td:nth-of-type(1)')

even_post_nums = soup.select('.list1 > td:nth-of-type(1)')

post1_num = odd_post_nums[0].text.strip()
post2_num = even_post_nums[0].text.strip()
post3_num = odd_post_nums[1].text.strip()

print(post1_num)
print(post2_num)
print(post3_num)
```

### 과제 3)
- 무작위 비밀번호 찾기

```python
import requests
import random
from string import ascii_lowercase
import time
from bs4 import BeautifulSoup

alpha_list = list(ascii_lowercase)
url = 'http://dgucn.iptime.org:9090/~aksen123-sms/hw-3.php'

password_temp = []

while True:
    password = 'b'
    for i in range(3):
        password += random.choice(alpha_list)
    
    print(password)
    
    if password in password_temp:
        continue
    
    time.sleep(0.5)
    
    r = requests.get(url, {'password' : password})
    html = r.text

    soup = BeautifulSoup(html, 'html.parser')
    if soup.find('p').get_text() == '틀림':
        password_temp.append(password)
    
    if soup.find('p').get_text() == '비밀번호 맞음':
        print('비밀번호는 :', password)
        break
    
    
    
```
